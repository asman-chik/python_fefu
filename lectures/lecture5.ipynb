### Менеджер пакетов pip

Основные команды:
- **pip help** - помощь по доступным командам
- **pip install package_name** - установка пакета(ов)
- **pip uninstall package_name** - удаление пакета(ов)
- **pip list** - список установленных пакетов
- **pip show package_name** - показывает информацию об установленном пакете
- **pip search** - поиск пакетов по имени
- **pip install -U** - обновление пакета(ов)
- **pip install --force-reinstall** - при обновлении, переустановить пакет, даже если он последней версии

Важные команды в сочетании с виртуальными средами окружения:
- **pip freeze > requirements.txt** - вывод всех установленных в системе (или в виртуальном окружении) пакетов с версиями в формате в сокращённом формате, удобного для дальнейшей установки (имя_пакета==версия), в файл requirements.txt
- **pip install -r > requirements.txt** - установка всех пакетов указанных в файле requirements.txt

### Виртуальные окружения

Виртуальное окружение - изолированный от общего системного Питона экзэмпляр дистрибутива Питона, в котором можно устанавливать и удалять пакеты различных версий, не затрагивая основной системный Питон.
<br><br>
Виртуальное окружение позволяет избежать конфликтов версий пакетов, версий Питона, изолировать приложение и все его зависимости (пакеты, необходимые для его работы.
<br><br>
Основные команды:
- **python -m venv venv-name** - создаст каталог venv-name, если он не существует, и также создаст директории внутри него, содержащие копию интерпретатора Python, стандартную библиотеку и различные вспомогательные файлы
- **venv-name\Scripts\activate.bat (Windows) или **source venv-name/bin/activate (Linux и MacOS)** - активация виртуального окружения, то есть теперь все команды, относящиеся к Питону и установленным в виртуальном окружении пакетам будут вызывать именно версии из виртуального окружения без необходимости прописывать явный путь до них или изменять переменные окружения
- **deactivate** - выход из виртуального окружения
<br><br>
Подробнее можно почитать [ЗДЕСЬ](https://pythoner.name/documentation/tutorial/venv)
