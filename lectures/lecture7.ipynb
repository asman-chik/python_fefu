### Нейронные сети, слои, нелинейность, параметры сети
Нейронная сеть - система соединенных и взаимодействующих между собой искусственных нейронов. Нейроны объединяются в слои.
![](https://cdn-images-1.medium.com/max/1000/1*eEKb2RxREV6-MtLz2DNWFQ.gif)

Входной слой - слой нейронов, выходы которых являются значениями входных данных. Количество нейронов в слое соответствует количеству признаков во входных данных.
<br><br>
Выходной слой - слой нейронов, выходы которых являются значением выходов для соответствующих входных данных. Количество нейронов соответствует количеству признаков в выходах.
<br><br>
Скрытые слои - слои нейронов, входы которых - выходы с предыдущих слоёв, а выходы - входы для следующих.
<br><br>
Параметры нейросети - все веса сети.
<br><br>
Гиперпараметры нейросети - количество слоёв, количество нейронов в слоях и т.д.

### Обучение нейросетей, целевая функция, прямой и обратный проходы
Прямой проход - расчёт значений выходного слоя на основе значений входного слоя через все скрытые слои. 
<br>
<br>
Целевая функция (функция потерь) - функция от обучающих выходных значений y0 и выходных значений нейросети y. Например, среднеквадратичное отклонение.
<br>
<br>
Обучение нейросети - процесс подбора параметров (весов) нейросети с целью миницизации функции потерь.
<br>
<br>
Обратный проход - дифферинцирование функции потерь по всем параметрам нейросети.


## Часть 2. Простейший пример нейронной сети

### Постановка задачи линейной регрессии
Для начала попробуем решить простую задачу: подобрать для функции y = kx + b параметры k и b при известном наборе значений x0 и y0.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n, k, b = 200, 2.5, 6
x0 = np.linspace(-2.0, 2.0, n)
y0 = x0 * k + b + np.random.normal(loc=0, scale=0.7, size=n)

fig, ax = plt.subplots(figsize=(8,3))
ax.scatter(x0, y0, marker='x'), ax.set_xlabel('x0'), ax.set_ylabel('y0');
```

### Модель и прямой проход
В качестве модели для решения задачи возьмём нейронную сеть из 1 входного и 1 выходного нейрона. Без функции активации, потому как нам нужна активация выходного нейрона в чистом виде.

![](https://i.stack.imgur.com/bI2Tm.gif)
![](https://i.stack.imgur.com/oapHD.gif)
Тогда прямой проход (предсказание y по входам x) будет следующим:
![](https://cdn-images-1.medium.com/max/1000/0*NOG_xxqN9f3-dKws.png)

### Функция потерь
В качестве функции потерь возьмём среднеквадратичную ошибку (Mean Square Error):
![](https://cdn-images-1.medium.com/max/1000/0*3tUgqNjtX102ruKT.png)
- i — номер точки из x0
- ŷ — модельное значение
- y — реальное значение
- m — количество точек во входном наборе данных
<br><br>
Можно переписать в виде:
![](https://cdn-images-1.medium.com/max/1000/0*3boprC82I9Aq2DS4.png)

```python
def mse(y_pred, y_real):
    samples_num = y_real.shape[0]
    mae_error = (1.0 / (2*samples_num)) * np.sum((y_pred - y_real)**2)
    return mae_error
```

### Обратный проход: обратное распространение ошибки и градиентный спуск
Для корректировки параметров модели используется так называемый метод обратного распространения ошибки (backpropagation). Он базируется на методе градиентного спуска (gradient descent).
![](https://cdn-images-1.medium.com/max/1000/1*iWRgqnhWm5CBt-0Pi6hzbg.gif)

Ещё раз приведём формулы расчёта ошибки (функции потерь) и градиенты для параметров w и b:
![](https://cdn-images-1.medium.com/max/1000/0*3boprC82I9Aq2DS4.png)
![](https://cdn-images-1.medium.com/max/1000/0*7Hv7sFAy5xOPdIQN.png)
![](https://cdn-images-1.medium.com/max/1000/0*2f2pJmQ3fQHrZ0GV.png)
Обновление значений параметров будем проводить по следующим формулам (где α - "скорость обучения"):
![](https://cdn-images-1.medium.com/max/1000/0*wOYOnCSQ9EJccJhy.png)

```python
# Зададим начальные параметры
w, b = 0, 0
iterations = 1000
learning_rate = 0.01
m = x0.shape[0]
errors = []

# Проведём обучение модели
for iteration in range(iterations):
    y = w * x0 + b
    dw = (1 / m) * np.sum((y - y0) * x0)
    db = (1 / m) * np.sum(y - y0)
    w = w - learning_rate * dw
    b = b - learning_rate * db
    errors.append(mse(y, y0))

# Отобразим результаты
fig, ax = plt.subplots(1, 2, figsize=(8, 2))
fig.tight_layout()
ax[0].plot(range(iterations), errors)
ax[0].set_xlabel('Iteration'), ax[0].set_ylabel('Error')
ax[1].scatter(x0, y0, marker='x')
ax[1].plot(x0, y, color='r', linewidth=3)
ax[1].set_xlabel('x0'), ax[1].set_ylabel('y0');    
```

### Линейная регрессия на PyTorch

```python
# Импорт необходимых библиотек
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Зададим те же самые входные данные
np.random.seed(42)
n, k, b = 200, 2.5, 6
x_train = np.linspace(-2.0, 2.0, n, dtype=np.float32)
y_train = x_train * k + b + np.random.normal(loc=0, scale=0.7, size=n)

# Гипер-параметры модели
input_size = 1
output_size = 1
num_epochs = 2000
learning_rate = 0.001

# Зададим нейросеть, используя специальный класс Linear, который реализует полносвязный слой
model = nn.Linear(input_size, output_size)

# Зададим функцияю потерь с помощью класса MSELoss и метод оптимизации - стохастический градиентный спуск
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)  
```

```python
# Конвертируем массивы numpy в тензоры - многомерные массивы PyTorch
inputs = torch.from_numpy(x_train.reshape(n,1))
targets = torch.from_numpy(y_train.reshape(n,1).astype(np.float32))
errors = []

# Обучим модель
for epoch in range(num_epochs):
    # Прямой проход
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    # Обратный проход и оптимизация. Обязательно обнулить градиенты!
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    errors.append(loss.item())
    
# Нарисуем графики
predicted = model(inputs)
fig, ax = plt.subplots(1,2,figsize=(8,3))
ax[0].plot(range(num_epochs), errors, label='Error')
ax[1].plot(x_train, y_train, 'ro', label='Original data')
ax[1].plot(x_train, predicted.detach().numpy(), label='Fitted line')
ax[0].legend(), ax[1].legend();
```

### Распознавание рукописных цифр: набор данных MNIST 
База данных MNIST (сокращение от «Modified National Institute of Standards and Technology») — объёмная база данных образцов рукописного написания цифр.
![](https://cdn-images-1.medium.com/max/1000/1*_91Ir1B68lDv_8kSRHa9bg.jpeg)

Идея архитектуры сети: изображения размером 28x28 = 784 пикселя, поэтому пусть у сети будет 784 входа.
![](https://cdn-images-1.medium.com/max/1000/1*wPvGyzWJRHO-yxbI57p2zg.png)

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms

# Гипер параметры
input_size = 784
hidden_size = 500
num_classes = 10
num_epochs = 5
batch_size = 100
learning_rate = 0.001

# Выбор устройства, на котором проводить расчёты
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Загрузим встроенный в PyTorch MNIST dataset 
train_dataset = torchvision.datasets.MNIST(root='data', train=True, transform=transforms.ToTensor(), download=True)
test_dataset = torchvision.datasets.MNIST(root='data', train=False, transform=transforms.ToTensor())

# Создадим специальный класс DataLoader, который будет подавать изображения порциями по batch_size
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# Зададим структуру нейронной сети с помощью класса, наследованного от nn.Module
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size) 
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, num_classes)  
    
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

model = NeuralNet(input_size, hidden_size, num_classes).to(device)

# Зададим функцию потерь и метод оптимизации
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)  

# Напечатаем структуру модели
print(model)
```

```python
# Обучим модель
total_step = len(train_loader)
for epoch in range(num_epochs):
    for i, (images, labels) in enumerate(train_loader):  
        # Представим изображения в виде вектора длиной 784
        images = images.reshape(-1, 28*28).to(device)
        labels = labels.to(device)
        
        # Прямой проход
        outputs = model(images)
        loss = criterion(outputs, labels)
        
        # Обратный проход и оптимизация
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # Печатаем информацию каждые 100 итераций
        if (i+1) % 300 == 0:
            print ('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}' 
                   .format(epoch+1, num_epochs, i+1, total_step, loss.item()))
```

```python
# Тестируем модель, для этого говорим PyTorch не считать градиенты
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in test_loader:
        images = images.reshape(-1, 28*28).to(device)
        labels = labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    print('Accuracy of the network on the 10000 test images: {} %'.format(100 * correct / total))
```

```python
# Посмотрим пример неверно определённых цифр
wrong = images[predicted != labels]
fig, ax = plt.subplots(1, 1)
array = wrong[0].reshape(28, 28).to('cpu').numpy()
ax.imshow(array, cmap='gray')
ax.set_title(f'Predicted: {predicted[predicted != labels][0]}\nGround Truth: {labels[predicted != labels][0]}');    
```
