### Делители числа
Напишите функцию, которая на вход получает число больше нуля, а возвращает список всех его делителей, т.е. список чисел, на которые входное число делится без остатка.

### Слова в обратном порядке
Напишите функцию, которая на вход получает строку со словами, а возвращает строку со словами в обратном порядке. Слово отделено с обеих сторон произвольным количеством пробелов или началом или концом строки.

### Количество дней в месяце
Напишите функцию, которая на вход получает два числа - месяц и год, а возвращает количество дней в месяце данного года. В феврале обычно 28 дней, а в високосном году 29. Високосный год - год, который делится без остатка на 4 и не делится при этом на 100 без остатка, или который делится только на 400 без остатка.

### Одинаковые символы в двух строках
Напишите функцию, которая на вход получает 2 строки, а возвращает строку, содержащую все символы, присутствующие в обеих входных строках.

### Посчитать количество вхождений в список
Напишите функцию, которая на вход получает произвольный список, а возвращает словарь, в котором ключи - элементы списка, а значения - количество вхождений элемента в исходный список.

### Оставить только слова длиной не менее n символов
Напишите функцию, которая на вход получает строку со словами, разделёнными произвольным количеством пробельных символов, и число n - минимальная длина искомых слов, а возвращает список слов-строк, длина которых не менее n.

### Баланс скобок
Напишите функцию, которая на вход получает строку с произвольными символами, а возвращает True, если количество открывающих круглых скобок “(“ совпадает с количеством закрывающих круглых скобок “)”, и False - в противном случае. При этом первой не может стоять закрывающая скобка.

### Разбить число на сумму порядков
Напишите функцию, которая на вход получает целое положительное число от 0 до 9999999, а возвращает строку с разложением числа в виде суммы порядков. Например: 5325 ➞ "5000 + 300 + 20 + 5".

### Валидация email
Напишите функцию, которая на вход получает строку, а возвращает True, если строка соответствует правилам написания email адреса, и False - в противном случае. Правила следующие: только один символ @, слева от него хотя бы один символ, справа от него хотя бы 2 символа, разделённых точкой.

### Количество букв и символов в строке
Напишите функцию, которая на вход получает строку из букв, цифр и других символов, а возвращает кортеж из двух чисел - количеством букв и количеством цифр в строке соответственно.

### Замаскировать строку
Напишите функцию, которая на вход получает произвольную строку, число n - количество символов для маскирования и строку из одного символа - символ маскирования, а возвращает исходную строку с замаскированными первыми n символами указанным символом.

### Две строки - анаграмма ли?
Напишите функцию, которая на вход получает две строки, а возвращает True, если строки являются анаграммами, и False - в противном случае. Одна строка анаграмма для другой, если состоит и тех же самых символов и в том же количестве, но не равна ей. При этом функция должна быть нечувствительна к регистру.
