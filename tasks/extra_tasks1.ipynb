## Дополнительные задания по курсу Python. По материалам занятий 1 и 2.

Постановка задачи в каждом задании достаточно общая, без конкретики и жёстких требований к реализации. Обычно так и ставятся все задачи, а пропработка деталей реализации остаётся за разработчиком.

### Задание 1. Генератор и валидатор пароля

Напишите две функции: для генерации пароля и валидации пароля. <br><br>
Генератор пароля должен выдавать строку-пароль, сгенерированную случайным образом. Для этого рекомендуется использовать модуль random (`import random`) и его функции `randint`, `choice` и т.д. Различные параметры генерации пароля (диапазон длины пароля, символы для составления пароля, требования надёжности пароля) придумайте и реализуйте сами. Нет предела совершенству (но есть минимальные требования и сроки).<br><br>
Валидатор пароля должен принимать строку-пароль и выдавать результат проверки: удовлетворяет или не удовлетворяет требованиям безопасности (длина, символы в разном регистре, спецсимволы и т.д.). Требования безопасности можно передавать в качестве аргументов функции.<br><br>
Проявите фантазию при написании функций. Можно реализовать самый минимальный функционал, а можно добавить дополнительных возможностей.

### Задание 2. Калькулятор доходности вклада в банке

Напишите функцию, которая расчитывает доходность вклада в банке в зависимости от условий вклада (сумма, процентная ставка, срок вклада, капитализация процентов). Функция должна принимать на вход параметры вклада, а выдавать сумму по окончании срока вклада.<br><br>
Варианты реализации те же, что и в задаче 1: проявите фантазию. Можно сделать несколько функций с разным набором воможностей.

### Задание 3. Крупье, раздающий игральные карты

Напишите функцию, которая выдаёт N карт из колоды. В выдаче не может быть двух одинаковых карт. Для красоты можно использовать символы мастей карт в кодировке юникод в сочетании с старшинством карты (2-10, J, Q, K, A). Результат работы функции - строка с картами.<br><br>
В качестве дополнительных возможностей можно добавить рамеры колод карт для разных игр  (52, 36, 28 и т.д.).

### Задание 4. Игра в кости

Напишите функцию, которая бросает кости (два кубика, значения от 1 до 6 на каждый кубик) и выдаёт строку, содержащую выпавшие значения и итоговый результат, который вычисляется по какой-нибудь формуле или таблице (например, разные числа - сумма, дубль - удвоенная сумма).
